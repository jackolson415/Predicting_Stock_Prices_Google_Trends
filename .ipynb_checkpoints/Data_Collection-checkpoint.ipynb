# Problem Statement

In this project I've set out to identify how using Google Trends data could inform the prediction of stock price changes. As a CPA I have studied the importance of the financial statments filed by public companies. These financial statments, of course, are used extensively by investors to predict stock price changes. In this project I hope to see if using Google Trends in conjunction with financial statement data could improve prediction ability.

Google Trends is an online tool that helps visualize trends in a term's searching popularity. The output of Google Trends is simply a time series of numbers between 1 and 100. A 100 indicates the time when the search was most popular over the time range selected. A 1 indicates when the search was least popular over that period.

Google Trends data goes back to January 1st 2004 for all search terms, even companies that were founded well after 2004. For example, here is the Google Trends output for the search term "PTON" since its IPO. PTON is the ticker symbol for popular exercise bike company Peloton.

Searches for "PTON" were most popular between 9/6/20-9/12/20. Below is the graph of the PTON price over the same period.

```python
Image('./Pictures/PTON Stock Snip.png', width=700)
```

As the stock chart above shows, PTON was trading at about 84 dollars as of 9.11.20, the period where Google Trends shows the maximum interest in searches for "PTON". In the months following the price of the stock rose by 104%.

Could investors have used the high search interest in PTON during September 2020 to predict such a large increase in the stock price? That is what this project will investigate.
